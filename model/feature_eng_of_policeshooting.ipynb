```
def f(row):
    if row['arms_category'] == 'Unarmed':
        val = 1
    elif row['arms_category'] == 'Other unusual objects':
        val = 1
    elif row['arms_category'] == 'Unknown':
        val = 1
    else:
        val = 0
    return val

data['target'] = data.apply(f, axis=1)
```

```
data['signs_of_mental_illness'] = data.signs_of_mental_illness.map(lambda a: 1 if a == True else 0)
data['body_camera'] = data.body_camera.map(lambda a: 1 if a == True else 0)
```

```
#encoding categorical columns
from sklearn.preprocessing import LabelEncoder
cat_cols = ['armed','city','state']
encoder = LabelEncoder()

#Apply label encoder to each column
encoded = data[cat_cols].apply(encoder.fit_transform)
```

```
#encoding categorical columns
from sklearn.preprocessing import LabelEncoder
cat_cols2 = ['manner_of_death','gender','race','signs_of_mental_illness','threat_level','flee','body_camera','arms_category']

#Apply label encoder to each column
encoded2 = pd.get_dummies(data[cat_cols2], drop_first=True)
```
